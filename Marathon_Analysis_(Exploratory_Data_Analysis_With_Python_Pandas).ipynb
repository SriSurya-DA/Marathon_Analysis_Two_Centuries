```python
# Now combaining both the filters

df[(df["Event distance/length"].isin(["50mi",'50km'])) & (df["Year of event"]==2020) & (df["Event name"].str.split('(').str.get(1).str.split(")").str.get(0).str.strip()=="USA")]
```

```python
# Now assigning an variable for the above filter as 'df2'

df2 = df[(df["Event distance/length"].isin(["50mi",'50km'])) & (df["Year of event"]==2020) & (df["Event name"].str.split('(').str.get(1).str.split(")").str.get(0).str.strip()=="USA")]
```

```python
df2.head(5)
```

```python
df2.shape
```

```python
# Now I like to remove the "USA" from the "Event Name"

df2["Event name"].str.split('(').str.get(0)
```

```python
# Now assigning this back to the "Event Name"
df2["Event name"] = df2["Event name"].str.split('(').str.get(0)
```

```python
df2 # success
```

```python
# Cleaning up th Athlete Age

# We having an Athlete age as an separate column but it is not correctly formated (M35, W40, MU37..)

# Hence I am calaculation (Year of the event [2020] - Athlete Year of birth)
```

```python
age = df2["Year of event"] - df2["Athlete year of birth"]
```

```python
# Removing 'h' from "Athlete performance"

df2["Athlete performance"] = df2["Athlete performance"].str.split(" ").str.get(0)
```

```python
df2.head(5) # Success
```

```python
df2["age"] = age # Above I have performed calculation (event year - Athlete birth year)
```

```python
df2.head()
```

```python
# Now dropping the unwanted column from df2 ["Athlete club, Athlete country, Athlete year of birth and Athlete age category"]
```

```python
df2 = df2.drop(["Athlete club", "Athlete country","Athlete year of birth","Athlete age category"], axis=1)
```

```python
# To drop the missing value

df2 = df2.dropna()
```

```python
# Rename colums

#Year of event
#Event dates
#Event name
#Event distance/length
#Event number of finishers
#Athlete performance
#Athlete gender
#Athlete average speed
#Athlete ID
```

```python
df2 = df2.rename(columns = {

    "Year of event": "year",
    "Event dates" : "race_day",
    "Event name": "race_name",
    "race_lenght": "race_length",
    "Event number of finishers": "race_number_of_finishers",
    "Athlete performance": "athlete_performance",
    "Athlete gender": "athlete_gender",
    "Athlete average speed": "athlete_average_speed",
    "Athlete ID": "athlete_id"

})
```
